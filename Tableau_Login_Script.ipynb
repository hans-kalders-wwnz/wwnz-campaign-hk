```
import tableauserverclient as TSC

# --- Prerequisites ---
# Before running this script, you must install the tableauserverclient library.
# You can do this by running the following command in your terminal:
# pip install tableauserverclient

# --- Configuration ---
# Replace these values with your specific Tableau Server/Cloud details.
# The server URL should look like 'https://10ax.online.tableau.com' for Tableau Cloud,
# or 'https://your-on-prem-server.com' for Tableau Server.
TABLEAU_SERVER_URL = "https://us-west-2b.online.tableau.com/"

# The site ID is the name of your site (e.g., 'your-site-name') or an empty string
# if you are on the default site.
TABLEAU_SITE_ID = "woolworthsnewzealand"

# Use the Personal Access Token name and secret that you created.
# These values are sensitive and should not be hardcoded in production code.
# The user's uploaded image shows "automation" as the token name and a long secret.
TOKEN_NAME = "auto3"
TOKEN_SECRET = "FxDyx0qbS2+hzryP6+6C0w==:Bs3pbY4PVCZGSYMWaW0yhIshmMR0q8br"

def sign_in_to_tableau():
    """
    Authenticates with Tableau Server or Tableau Cloud using a Personal Access Token.
    """
    try:
        # Create an authentication object using the Personal Access Token.
        tableau_auth = TSC.PersonalAccessTokenAuth(
            token_name=TOKEN_NAME,
            personal_access_token=TOKEN_SECRET,
            site_id=TABLEAU_SITE_ID
        )

        # Create an instance of the Tableau Server client.
        server = TSC.Server(TABLEAU_SERVER_URL, use_server_version=True)

        # Sign in to the server. The 'with' statement ensures the connection is
        # properly closed even if an error occurs.
        with server.auth.sign_in(tableau_auth):
            print(f"Successfully signed in to Tableau Server at {server.baseurl}!")
            print(f"Server version: {server.version}")
            # print(f"Current user ID: {server.auth.user_id}")
            # You can now perform other operations, such as getting workbooks or views.
            # Example: Get a list of all workbooks on the site.
            all_workbooks, pagination_item = server.workbooks.get()
            print(f"Found {len(all_workbooks)} workbooks on the site.")
            print("Workbooks found:")
            for workbook in all_workbooks:
                print(f"- {workbook.name}")


    except TSC.ServerResponseError as e:
        print(f"Error signing in to Tableau: {e}")
        print("Please check your server URL, site ID, token name, and secret.")
        print("If you are on Tableau Cloud, ensure your site ID is correct.")

if __name__ == "__main__":
    sign_in_to_tableau()
```
